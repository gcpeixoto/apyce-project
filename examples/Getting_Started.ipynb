### Import APyCE module
The APyCE module has two classes, however, only the Model class is used by the user.

```python
import apyce as ap
```

### Using
#### Reading a file
When the user executes the line below, the file is loaded and all keywords are readed.

```python
G = ap.grid.Grid('Data/PSY.grdecl', 'eclipse')
```

#### Printing the structure
If the user needs to visualize the data, he can do it with the line below.

```python
print(G)
```

#### Processing the grid

```python
G.process_grid()
```

#### Read data from another file

```python
#G.load_cell_data('./Data/dome_Temperature.txt', 'TEMP')
```

#### Saving the grid in VTU format

```python
G.export_data()
```

#### Plotting the grid with PyVista

```python
G.plot_grid(filename='Results/PSY.vtu', lighting=False, property='PORO', show_edges=True, specular=0.0,
            specular_power=0.0, show_scalar_bar=True, cmap='viridis')
```
